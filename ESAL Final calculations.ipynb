```python
#Traffic for lean  season
#ADT for non-harvesting season

#Non-Motorised vehicles 
cycle=5
cyclerickshaw=2
animaldrawnvehicles=3

#Motorised vehicles
carjeepvan3W=4
twoW=5
LCV=5

truckladen=6
truckunladen=1
truckoverloaded=3

busladen=4
busunladen=3
busoverloaded=4


tractorladen=2
tractorunladen=3
tractoroverloaded=2

ADT= (cycle + cyclerickshaw+animaldrawnvehicles+carjeepvan3W+twoW+LCV+truckladen+
      truckunladen+truckoverloaded+
      busladen+busunladen+busoverloaded+tractorladen+tractorunladen+tractoroverloaded)

print(ADT, "is ADT")
AADT=(ADT + ((1.2*ADT*75)/365))
print(AADT, "is AADT")

#AADT after 2 years
AADT2=AADT*1.06*1.06
print(AADT2, "is AADT after 2 years")

#truckladen after 2 years
truckladen2=(AADT2*truckladen)/ADT
print(truckladen2, "are trucks laden after two years")

#truckunladen after 2 years
truckunladen2=(AADT2*truckunladen)/ADT
print(truckunladen2, "are trucks unladen after two years")

#truckoverloaded after 2 years
truckoverloaded2=(AADT2*truckoverloaded)/ADT
print(truckoverloaded2, "are trucks overloaded after two years")

#busladen after 2 years
busladen2=(AADT2*busladen)/ADT
print(busladen2, "are bus laden after two years")

#busladen after 2 years
busunladen2=(AADT2*busunladen)/ADT
print(busunladen2, "are bus unladen after two years")

#busoverloaded after 2 years
busoverloaded2=(AADT2*busoverloaded)/ADT
print(busoverloaded2, "are bus overloaded after two years")

#tractorladen after 2 years
tractorladen2=(AADT2*tractorladen)/ADT
print(tractorladen2, "are tractors laden after two years")

#tractor unladen after 2 years
tractorunladen2=(AADT2*tractorunladen)/ADT
print(tractorunladen2, "are tractor unladen after two years")

#tractoroverloaded after 2 years
tractoroverloaded2=(AADT2*tractoroverloaded)/ADT
print(tractoroverloaded2, "are tractor overloaded after two years")

#ESAL for HCV_truck

total_laden=truckoverloaded2+truckladen2
total_overloaded=truckoverloaded2
overloaded_percent=(total_overloaded/total_laden)
laden_percent=1-(overloaded_percent)
print(overloaded_percent, "is truck overloaded percent")
print(laden_percent, "is truck laden percent")


ESAL_HCVtruck=((((overloaded_percent)*(total_laden*5.35))+((laden_percent)*(total_laden*2.58)))+(0.31*(truckunladen2)))
print(ESAL_HCVtruck, "is HCV ESALtruck")

#ESAL for HCV_bus

total_laden1=busoverloaded2+busladen2
total_overloaded1=busoverloaded2
overloaded_percent1=(total_overloaded1/total_laden1)
laden_percent1=1-(overloaded_percent1)
print(overloaded_percent1, "is bus overloaded1 percent")
print(laden_percent1, "is bus laden1 percent")

ESAL_HCVbus=((((overloaded_percent1)*(total_laden1*5.35))+((laden_percent1)*(total_laden1*2.58)))+(0.31*(busunladen2)))
print(ESAL_HCVbus, "is HCV ESALbus")


#ESAL for MCV

total_laden2=tractoroverloaded2+tractorladen2
total_overloaded2=tractoroverloaded2
overloaded_percent2=(total_overloaded2/total_laden2)
laden_percent2=1-(overloaded_percent2)
print(overloaded_percent2, "is tractor overloaded2 percent")
print(laden_percent2, "is tractor laden2 percent")

ESAL_MCV_tractor=((((overloaded_percent2)*(total_laden2*0.65))+((laden_percent2)*(total_laden2*0.31)))+(0.02*(tractorunladen2)))
print(ESAL_MCV_tractor, "is MCV ESAL_tractor")


#Total ESAL
ESAL=ESAL_HCVtruck+ESAL_HCVbus+ESAL_MCV_tractor
print(ESAL)

#ESAL Cummulative
ESAL_cum=4811*ESAL
print(ESAL_cum, "is the cummulative ESAL")

if ESAL_cum<=30000:
    print("T1 Category")
elif ESAL_cum<=60000:
    print("T2 Category")
elif ESAL_cum<=100000:
    print("T3 Category")
elif ESAL_cum<=200000:
    print("T4 Category")
elif ESAL_cum<=300000:
    print("T5 Category")
elif ESAL_cum<=600000:
    print("T6 Category")
elif ESAL_cum<=1000000:
    print("T7 Category")
elif ESAL_cum<=1500000:
    print("T8 Category")
elif ESAL_cum<=2000000:
    print("T9 Category")
    
else:
    print("Go For IRC 37 Design")
```
